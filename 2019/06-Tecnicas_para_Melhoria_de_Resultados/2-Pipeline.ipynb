# Pipeline

```python
from sklearn import datasets
from sklearn import preprocessing
from sklearn import pipeline
from sklearn import model_selection
from sklearn import neighbors 
from sklearn import linear_model
from sklearn import svm
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y = datasets.load_breast_cancer(return_X_y=True)
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, random_state=42)
```

```python
model = svm.SVC(gamma='auto')
model_selection.cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy')
```

```python
s = preprocessing.StandardScaler()
model2 = svm.SVC(gamma='auto')
ppl = pipeline.Pipeline([('transformer', s), ('estimator', model2)])
model_selection.cross_val_score(ppl, X_train, y_train, cv=3, scoring='accuracy')
```
